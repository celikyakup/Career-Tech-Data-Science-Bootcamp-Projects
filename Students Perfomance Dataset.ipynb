```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sp=pd.read_csv("StudentsPerformance.csv")
sp.loc[:20]
```

```python
sp.columns
```

```python
sp.shape
```

```python
def getByGender(gender):
    
    return sp[sp["gender"]==gender] 
# datadaki gender sutünündaki veriler string değerler olduğu için def komutunun içine string olarak yazılması gereklidir.
```

```python
getByGender("male")
```

```python
#Gender sutünunu integer ifadeye çevirerek de aynı işlemi yapabiliriz.
sp_gender_int=sp.copy()
```

```python
# Female =0 ,male=1
sp_gender_int["gender"].replace("female",0,regex=True,inplace=True)
sp_gender_int["gender"].replace("male",1,regex=True,inplace=True)
sp_gender_int["gender"]
```

```python
sp["gender"].replace(0,"female",regex=True,inplace=True)
sp["gender"].replace(1,"male",regex=True,inplace=True)
sp
```

```python
def getByGenderStr(gender):
    return sp_gender_int[sp_gender_int["gender"]==gender]
getByGenderStr(0)
```

```python
def countByParentEducation():
    
    return sp["parental level of education"].value_counts(sort=True)

countByParentEducation()
```

```python
def getStatsByPrep():
    
    return sp.groupby("test preparation course")[["math score","reading score","writing score"]].agg([min,max,np.mean])
    
getStatsByPrep()
```

```python
sp.pivot_table(values=["math score","reading score"],index="race/ethnicity",columns="gender",aggfunc=[min,max,np.mean])
```

```python
# plt.scatter komutuyla yapabildiğim hali bu şekildedir.

sp_gender_int2=sp.copy()
sp_gender_int2["race/ethnicity"].replace("group A",0,regex=True,inplace=True)
sp_gender_int2["race/ethnicity"].replace("group B",1,regex=True,inplace=True)
sp_gender_int2["race/ethnicity"].replace("group C",2,regex=True,inplace=True)
sp_gender_int2["race/ethnicity"].replace("group D",3,regex=True,inplace=True)
sp_gender_int2["race/ethnicity"].replace("group E",4,regex=True,inplace=True)

sp_gender_int2["lunch"].replace("standard",0,regex=True,inplace=True)
sp_gender_int2["lunch"].replace("free/reduced",1,regex=True,inplace=True)

sp_gender_int2["test preparation course"].replace("none",0,regex=True,inplace=True)
sp_gender_int2["test preparation course"].replace("completed",1,regex=True,inplace=True)

x=sp["math score"]
y=sp["reading score"]
z=sp_gender_int["gender"]
q=sp_gender_int2["race/ethnicity"]
e=sp_gender_int2["lunch"]
f=sp_gender_int2["test preparation course"]
       
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 12))
b=ax[0][0].scatter(x,y,c=z)
ax[0][0].set(xlabel="math score",ylabel="reading score")
legend = ax[0][0].legend(*b.legend_elements(),loc="upper left", title="gender")
ax[0][0].add_artist(legend)

c=ax[0][1].scatter(x,y,c=q)
ax[0][1].set( xlabel="math score", ylabel="reading score")
legend1 = ax[0][1].legend(*c.legend_elements(),loc="upper left", title="race/ethniticy")
ax[0][1].add_artist(legend1)

d=ax[1][0].scatter(x,y,c=e)
ax[1][0].set( xlabel="math score", ylabel="reading score")
legend2 = ax[1][0].legend(*d.legend_elements(),loc="upper left", title="lunch")
ax[1][0].add_artist(legend2)

g=ax[1][1].scatter(x,y,c=f)
ax[1][1].set( xlabel="math score", ylabel="reading score")
legend3 = ax[1][1].legend(*g.legend_elements(),loc="upper left", title="test preparation course")
ax[1][1].add_artist(legend3);
```
