```python
#Data setlere yeni feature ekliyoruz.
# Humidity*light/Temperature
df1["(h*l)/t"]=df1["Humidity"]*df1["Light"]/df1["Temperature"]
df2["(h*l)/t"]=df2["Humidity"]*df2["Light"]/df2["Temperature"]

#Aynı işlemi dt3 data seti içinde yapıyoruz.
df3["(h*l)/t"]=df3["Humidity"]*df3["Light"]/df3["Temperature"]
```

```python
#Veri seti 1 ve veri seti 2 birbirlerinin devamı niteliğinde olduğu için verileri birleştirmiştik.
#Bu sebeple combine veri setiyle model oluşturuacağız
#Veri seti 3 ile bu 2 veri setinden oluşturulan modeli test edeceğiz.
#Train ve test değişkenlerini oluşturma.
# Train
X_train=combine.drop(columns=["date","Occupancy"]).values
y_train=combine["Occupancy"]

#Test
X_test=df3.drop(columns=["date","Occupancy"]).values
y_test=df3["Occupancy"]

#Değişkenlerimizin shapelerini kontrol ediyoruz.
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train.var(),X_test.var()
```

```python
#X_train ve X_test değişkenlerimizin değerleri anlaşılır olmadığı için normalize etmek için StandardScaler işlemi yapacağız.

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#Scaler ettikten sonra X_train,X_test değişkenlerimiz normalize hale gelmiştir.
X_test,X_train
```

```python
# Modeli eğitmek için birden fazla yaklaşım ile deniyoruz. En iyi sonuç verenler arasından seçim yapıyoruz.

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier

#burada modelleri bir listenin içerisine alıp parametreleri ile beraber tanımlıyoruz.
models = []
models.append(('RandomForestClassifier', RandomForestClassifier()))
models.append(('Decision Tree ',DecisionTreeClassifier())) 
models.append(('Logistic Regression', LogisticRegression()))
models.append(('K-NN', KNeighborsClassifier()))
models.append(('SVM', SVC()))
models.append(('Gradient Boosting Classifier', GradientBoostingClassifier()))
models.append(('AdaBoostClassifier', AdaBoostClassifier()))
models.append(('BaggingClassifier', BaggingClassifier()))

#burada for döngüsü vasıtasıyla tek tek bütün modelleri deneyerek sonuçları karşılaştırıyoruz. 
for name, model in models:
    model = model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    from sklearn import metrics

    print("%s -> ACC: %%%.2f" % (name,metrics.accuracy_score(y_test, y_pred)*100))
```

```python
# Çıkan sonuçlara bakarak en yüksek skor-uygunluk derecesine bakarak model seçimi yapıyoruz. 
#Modelimiz sınıflandırma problemi üzerine olduğu için işlemlerimize Random Forest Classifier modeli ile devam ediyoruz.

#Seçtiğimiz modelden en iyi sonucu alabilmek için hyperparametre tuning yapıyoruz.
rf = RandomForestClassifier(max_depth=10, n_estimators=100, max_samples=0.6,
                             max_features=.7, random_state=42,verbose=0,oob_score=True)

rf.fit(X=X_train,y=y_train)

#Modelimizi test edeceğimiz verimizi giriyoruz ve doğruluk oranına bakıyoruz.
y_pred=rf.predict(X_test)
print("ACC:",accuracy_score(y_true=y_test,y_pred=y_pred))

#Modelimizin accuracy score değeri yüksek çıkmıştır. Ama bu modeli kabul etmek için tek başına yeterli değildir.
```

```python
# Confusion matrise bakıldığında modelimiz 9 yerde insan yoku var olarak algılamıştır.
#Precision kesinlik değeridir yüksek olması model için önemlidir. 
#Modelin doğru tahminlediği değerlerin kaçının doğru olduğunu bildiriyor.
#Recall doğru tahminlediği işlemlerinin kaçının doğru olduğunu bildiriyor.
#F1-score ise bu iki parametrenin harmonik ortalamasıdır.

from sklearn.metrics import confusion_matrixg

cnfm = confusion_matrix(y_true=y_test, y_pred=y_pred)
plt.figure(dpi=120)
sns.heatmap(cnfm, annot=True, cmap="viridis")
print(classification_report(y_true=y_test, y_pred=y_pred));
```
