```python
# Verinin içerisindeki 20 dakikadan kısa filmleri seçebilen, ve yeni bir sütunda filmlerin kısa film olup olmaıdğını belirten kodu

film_mi=nf[nf["type"]=="Movie"]
film_mi["duration"].unique()
```

```python
film_mi["duration"].replace(" min","",regex=True,inplace=True)
film_mi["duration"].unique()
film_mi["duration"] =pd.to_numeric(film_mi["duration"])
```

```python
film_mi_median=film_mi["duration"].median()
film_mi_median
```

```python
film_mi["duration"].fillna(film_mi_median,inplace=True)
film_mi["duration"].isnull().sum()
```

```python
# function that converts not a number values to desired data 
copy_nf=nf.copy()

def replaceNan(pColumn,pValue):
    
    return copy_nf[pColumn].fillna(pValue,inplace=True),display(copy_nf)
    
```

```python
replaceNan("country","Country")
```

```python
#input: director name , output: director's TV shows and movies
def getByDirector(name):
    ax=nf[nf["director"]==name]
   
    if ax.size>0:
        
        return ax[["director","title"]]                     
                                     
    else:
        print("The series or movies of this director could be not found.")
```

```python
getByDirector("Julien Leclercq")
```

```python
#count catplot
with sns.axes_style("white"):
    g = sns.catplot(x="release_year", data=nf, aspect=2.5, kind="count", hue="type",
                    
                    palette=sns.color_palette("pastel")
                    )
    
    g.set_xticklabels(step=5)
    g.set_titles("Release Count")
```
